# CDF(Cummulative Distributed Function)

```python
One1 = df.loc[df["Surv_status"] == 1]
Two2 = df.loc[df["Surv_status"] == 2]
```

```python
label = ["Class 1 : PDF", "Class 1 : CDF", "Class 2 : PDF", "Class 1 : CDF"]
counts, bin_edges = np.histogram(One1["Age"], bins=10, density = True)
```

```python
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.title("PDF and CDF for Age")
plt.xlabel("Age")
plt.ylabel("% of Persons")
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
counts, bin_edges = np.histogram(Two2["Age"], bins=10, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(label)

plt.show()
```

```python
label = ["Class 1 : PDF", "Class 1 : CDF", "Class 2 : PDF", "Class 1 : CDF"]
counts, bin_edges = np.histogram(One1["Op_year"], bins=10, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
counts, bin_edges = np.histogram(Two2["Op_year"], bins=10, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.title("PDF and CDF for Op_year")
plt.xlabel("Op_year")
plt.ylabel("% of Persons")
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(label)

plt.show();
```

```python
label = ["Class 1 : PDF", "Class 1 : CDF", "Class 2 : PDF", "Class 1 : CDF"]
counts, bin_edges = np.histogram(One1["ax_node"], bins=10, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
counts, bin_edges = np.histogram(Two2["ax_node"], bins=10, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.title("PDF and CDF for ax_node")
plt.xlabel("ax_node")
plt.ylabel("% of Persons")
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.show();
```
